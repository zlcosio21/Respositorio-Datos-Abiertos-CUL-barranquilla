# **Tablas de frecuencias**

```python
frequency_table_Artritis=data['Artritis'].value_counts()
frequency_table_Diabetes=data['Diabetes'].value_counts()
frequency_table_Hipertensión=data['Hipertensión'].value_counts()
frequency_table_EPOC=data['EPOC'].value_counts()
frequency_table_Asma=data['Asma'].value_counts()
frequency_table_Insuficiencia_cardiaca=data['Insuficiencia cardiaca'].value_counts()
frequency_table_Cáncer=data['Cáncer'].value_counts()

resultados = pd.DataFrame({
    'Enfermedad': ['Artritis', 'Diabetes', 'Hipertensión', 'EPOC', 'Asma', 'Insuficiencia cardiaca', 'Cáncer'],
    '1 = Presenta': [
        frequency_table_Artritis.get(1, 0),
        frequency_table_Diabetes.get(1, 0),
        frequency_table_Hipertensión.get(1, 0),
        frequency_table_EPOC.get(1, 0),
        frequency_table_Asma.get(1, 0),
        frequency_table_Insuficiencia_cardiaca.get(1, 0),
        frequency_table_Cáncer.get(1, 0)
    ],
    '2 = No Presenta': [
        frequency_table_Artritis.get(2, 0),
        frequency_table_Diabetes.get(2, 0),
        frequency_table_Hipertensión.get(2, 0),
        frequency_table_EPOC.get(2, 0),
        frequency_table_Asma.get(2, 0),
        frequency_table_Insuficiencia_cardiaca.get(2, 0),
        frequency_table_Cáncer.get(2, 0)
    ]
})

resultados
```

Por lo anterior, se puede apreciar que, las enfermedades mas relevantes son

*   Diabetes
*   Hipertensión

Resultado de las tablas de frecuencias de las principales enfermedades registradas en el conjunto de datos.


```python


# Crear una figura con subgráficos
fig, axes = plt.subplots(3, 3, figsize=(15, 10))
fig.suptitle('Diagramas Circulares de Frecuencia por Enfermedad 1 = presenta y 2 = no presenta ', fontsize=16)

# Lista de enfermedades y sus tablas de frecuencia
enfermedades = [
    ('Artritis', frequency_table_Artritis),
    ('Diabetes', frequency_table_Diabetes),
    ('Hipertensión', frequency_table_Hipertensión),
    ('EPOC', frequency_table_EPOC),
    ('Asma', frequency_table_Asma),
    ('Insuficiencia cardiaca', frequency_table_Insuficiencia_cardiaca),
    ('Cáncer', frequency_table_Cáncer)
]

# Iterar sobre las enfermedades y los ejes para graficar
for ax, (enfermedad, freq_table) in zip(axes.flatten(), enfermedades):
    ax.pie(freq_table, labels=freq_table.index, autopct='%1.1f%%', startangle=90)
    ax.set_title(enfermedad)

# Ajustar el layout para que no se superpongan los subgráficos
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

Por lo anterior, se puede apreciar que, las enfermedades mas relevantes son

*   Diabetes
*   Hipertensión




# **Correlacion**

```python

numeric_data = data.select_dtypes(include=['int64', 'float64'])


correlation_matrix = numeric_data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt=".2f", linewidths=0.5)


plt.title("Matriz de Correlación de Variables Numéricas")


plt.show()
```
